```python
### Notebook to create a simple chatbot using gradio and OpenAI API

import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr

import warnings
warnings.filterwarnings('ignore')
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
```

```python
openai = OpenAI()
```

```python
system_message = "You are a powerful large language model which provides detailed information"
```

```python
def chat(message, history):  
    '''History is a list of dictionaries in an OpenAI-style message format. See the figure below
      to see how the history is stored. The history can be renamed as anything.'''

    return f"You said {message} and the history is {history}, [end of history]"

gr.ChatInterface(fn=chat).launch()
```

```python
## lets create a different function:

def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history] ## get only the role and content from history
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}] ## concatenate the list of dictionaries
    ## the system message, plus the entire history, plus the current message
    response = openai.chat.completions.create(model='gpt-4.1-mini', messages=messages)
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat).launch()
```
