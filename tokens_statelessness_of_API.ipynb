##### In this document we will convert words to tokens and revert back the tokens to words. We will also look at the statelessness of API calls, that is, every new API call is a new state.

```python
import tiktoken ## tiktoken does not work for all models

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Tell me about the tokens that are in the word microscope")
print(tokens)
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

```python
## lets decode the token 77
decoded = encoding.decode([77])
print(f"The corresponding token for the token number 77 is {decoded}")
```

#### Statelessness of API

```python
from openai import OpenAI
openai = OpenAI()
```

```python
messages = [
    {"role": "system", "content": "You are highly advanced LLM Model"},
    {"role": "user", "content": "Tell me who invented calculus"}
   ]
```

```python
response = openai.chat.completions.create(model="gpt-5.4-nano",  messages=messages)
text = response.choices[0].message.content
print(text)
```

```python
messages = [
    {"role": "system", "content": "You are highly advanced LLM Model"},
    {"role": "user", "content": "Tell me more about one of the scientist you mentioned"}
    ]
```

```python
## The API call wont be able to remember the last API call. The prompts should be written in a way, 
## that the API actually remembers the promts. Such as multiple prompts in a single call.

response = openai.chat.completions.create(model="gpt-5.4-nano",  messages=messages)
text = response.choices[0].message.content
print(text)
```
