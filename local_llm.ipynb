##### The API key in the following code can be set as anything since this is run locally. It is only a placeholder. The localhost refers to the local computer. The number 11434 is the port number. Bascially it means to send requests to the Ollama server running on the local computer. The port number tells the incoming network traffic which application or service should receive the data.

```python

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
response = ollama.chat.completions.create(model="llama3.2:1b", messages=[{"role": "user", "content": "Tell me a fact about LLM"}])

text = response.choices[0].message.content ## to print on a different line
print(text)
```
