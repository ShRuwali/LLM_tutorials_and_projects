#### A simple script to make use of OpenAI API

```python
import os
from dotenv import load_dotenv
```

```python
from openai import OpenAI
openai = OpenAI()

response = openai.responses.create(
    model="gpt-5-nano",                                             ## the model 
    input="Write a one sentence about LLMs."
)

print(response.output_text)
```

```python
# Define our system prompt

system_prompt = """
You are a Chess grandmaster playing against a 2000 rated player in Chess. You are playing as Black. There is no time limit. Provide me the best moves
to play and give me the reasoning as well but make sure the reasoning is less than 50 words. If the opponent makes an incorrect move then make sure to 
let me know as well. 
"""
```

```python
# Define our user prompt

user_prompt = """
The move the opponent has played is
"""
```

```python
def user_msg(Move):
    return [{"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt+ Move}
            ]
```

```python
input = 'd4' ## the move made by the opponent. You can simply create
            ## and input the move by the opponent code as well.
Move = input
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=user_msg(Move))
text = response.choices[0].message.content ## get the response
print(text) ## prints text without \n 
```

```python
## Another simple example
## Define the system prompt

system_prompt = """
You are a powerful LLM that provides detailed explanation of the output
"""

# Define our user prompt

user_prompt = """                                 ## This can be setup later as well
Tell me about LLMs in 1 sentence
"""
```
