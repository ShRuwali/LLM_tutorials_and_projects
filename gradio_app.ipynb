```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr 
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
```

```python
openai = OpenAI()
```

```python

system_message = "You are a powerful LLM"

def message_gpt(prompt):
    messages = [{"role": "system", "content": system_message}, {"role": "user", "content": prompt}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
message_gpt("What is today's date?") ## it will provide incorrect date
```

```python
# here's a simple function

def capitalize(text):
    print(f"Capitalize has been called with input {text}")
    return text.upper()
```

```python
capitalize("hello")
```

```python
gr.Interface(fn=capitalize, inputs="textbox", outputs="textbox", flagging_mode="never").launch()
```

```python
## now lets use the API


message_input = gr.Textbox(label="Your message:", info="Enter a message for GPT-4.1-mini", lines=7)
message_output = gr.Textbox(label="Response:", lines=8)

view = gr.Interface(
    fn=message_gpt,
    title="Chat with GPT", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=["hello", "Tell me an interesting fact"], 
    flagging_mode="never"
    )
view.launch()
```

```python
## lets print the output in a markdown format

system_message = "You are a powerful LLM"

message_input = gr.Textbox(label="Your message:", info="Enter a message for GPT-4.1-mini", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=message_gpt,
    title="Chat with OpenAI API", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Tell me about different types of LLM",
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
## instead of presenting the output as a bulk, we can also stream the results

def stream_gpt(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = openai.chat.completions.create(
        model='gpt-4.1-mini',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
import warnings
warnings.filterwarnings('ignore') ## ignore the jupyter warning

message_input = gr.Textbox(label="Your message:", info="Enter a message for GPT-4.1-mini", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_gpt,
    title="Chat with OpenAI API using Gradio App", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Explain to me about LLMs"
        ], 
    flagging_mode="never"
    )
view.launch()
```
